```python
class Node(object):
    def __init__(self, data):
        self.data = data
        self.children = []

    def addChild(self, obj):
        self.children.append(obj)
```

```python
p = Node(5)
q = Node(6)
r = Node(7)
p.addChild(q)
p.addChild(r)

print("p -> ", p.data)
for c in p.children:
    print("c: ", c.data)
```
